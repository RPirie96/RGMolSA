# Riemannian Geometry for Molecular Surface Approximation (RGMolSA)

**Contributing Authors:** Stuart J. Hall, Daniel J. Cole, Rachael M. E. Pirie

This notebook accompanies our paper detailing a novel method for molecular surface approximation based on Riemannian Geometry. The 9-value descriptor produced in this notebook represents the first 9 eigenvalues of the spectrum of the Laplacian associated with the surface. 

This notebook can be used to replicate the small screen detailed in the paper, comparing the shape of a series of PDE5 inhibitors known to have similar shape. Note that the method is currently still under development, and this notebook is not designed to be used for full virtual screens. 

```python
import sys
import os
sys.path.append('/home/b9046648/RGMolSA/scripts')  # replace with path to python scripts

from utils import get_score
from get_descriptor import get_descriptor

from rdkit.Chem import PandasTools, Draw, MolToMolBlock, rdMolAlign
import py3Dmol

root = '/home/b9046648/RGMolSA/data'  # specify directory containing data
```

## Single Conformer

The following cells compute the descriptor and calculate the similarity of a single conformer for each of Sildenafil, Vardenafil and Tadalafil. 

```python
# load data
data = PandasTools.LoadSDF(os.path.join(root, 'SVT.sdf'))  # replace with path to data
mols = list(data['ROMol'])
names = ['Sildenafil', 'Vardenafil', 'Tadalafil']
```

```python
# visualise molecules
Draw.MolsToGridImage(mols)
```

```python
# get the shape descriptor for each molecule in the list
descriptor = [get_descriptor(mol) for mol in mols]

print(descriptor)
```

```python
# compute inverse Bray-Curtis distance between molecules
# if a molecule id is supplied (in this case i) 'self' is returned if the molecule is compared to itself
scores = []
for i, des1 in enumerate(descriptor):
    s = [(get_score(des1, des2, i, j)) for j, des2 in enumerate(descriptor)]
    print('{0} Scores: {1}'.format(names[i], s))
    scores.append(s)

#print(scores)
```

## Multi-Conformer Similarity

The following cells compute the similarity between different conformers of the same molecule. The supplied datasets provide 11 random and 11 low energy conformers for Sildenafil, Vardenafil and Tadalafil, generated using RDKit. 

### Sildenafil Random Conformers

```python
# load data
data = PandasTools.LoadSDF(os.path.join(root, 'sildenafil_confs_10random.sdf')) 
sildenafil_random = list(data['ROMol'])

# visualise overlays

ref = sildenafil_random[0]

# draw molecules
p = py3Dmol.view(width=800,height=800)
p.enableContextMenu(contextMenuEnabled = True)
p.addModel(MolToMolBlock(ref), 'sdf')

for mol in sildenafil_random[1:]:
    rdMolAlign.AlignMol(ref, mol)
    p.addModel(MolToMolBlock(mol), 'sdf')

p.setStyle({'stick':{}})
p.setBackgroundColor('#FFFFFF')
p.zoomTo()
p.show()
```

```python
# get descriptors
sildenafil_random_descriptors = [get_descriptor(mol) for mol in sildenafil_random]  

# compute similarity
scores_sildenafil_random = []
for i, des1 in enumerate(sildenafil_random_descriptors):
    s = [get_score(des1, des2, i, j) for j, des2 in enumerate(sildenafil_random_descriptors)]
    scores_sildenafil_random.append(s)

print(scores_sildenafil_random)
```

## Potential Decoys

To check the method does not just assign high similarity to all molecules, 4 potential decoys were selected from the DrugBank Database:
- Arginine: lower MW and smaller size
- Lymecycline: higher MW but potential partial structural match with 4-ring motif (especially if one of the ring atoms is used as the base sphere)
- Diflorasone: similar MW and 4-ring motif, different therapeutic target/indication
- S-octylglutathione: similar MW and branching in molecule centre, no rings

### Arginine

```python
# load data
arginine = PandasTools.LoadSDF(os.path.join(root, 'Arginine.sdf'))  # replace with path to data
arginine = arginine.iloc[0]['ROMol']

# draw structure
Draw.MolToImage(arginine)
```

```python
# get descriptors
arginine_descriptor = get_descriptor(arginine)

# compute inverse Bray-Curtis distance between molecules
for i, des1 in enumerate(descriptor):
    s = get_score(des1, arginine_descriptor) 
    print('Similarity between Arginine and {0}: {1}'.format(names[i], s))
```
